```python
from esp import *
from ccs175 import *
import matplotlib.pyplot as plt
driver = esp()
spec = spectrometer()
time.sleep(1)
'''units = driver.QueryUnits()
print('units: ',units)
speed = driver.QuerySpeed()
print('speed; ',speed)'''
pos = 0
spec.set_integration_time(2e-3)
start_time = time.time()
time_trace = []
pos_trace = []
intensity = []
wav = []
while pos < 4:
    driver.setpos(pos)
    curr_pos = driver.getpos()
    intensity.append(spec.get_spectrum())
    wav.append(spec.get_wavelengths())
    #print(curr_pos)
    pos_trace.append(pos)
    time_trace.append(time.time()-start_time)
    pos+=.5
    time.sleep(.2)

    #time.sleep(.1)
plt.plot(time_trace, pos_trace)
plt.xlabel('Time in secondss')
plt.ylabel('Position in microns')
plt.show()
plt.plot(wav, intensity)
plt.xlabel('Wavelength [nm]')
plt.ylabel('Intensity [a.u.]')
plt.ylim(-0.1, .05)
plt.show()
driver.close()
spec.close()
```
